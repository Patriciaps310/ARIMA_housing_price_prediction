```python
#plotly graph for seasonal decomposition from tomron

def plot_seasonal_decompose(result:DecomposeResult, dates:pd.Series=None, title:str="Seasonal Decomposition"):
    x_values = dates if dates is not None else np.arange(len(result.observed))
    return (
        make_subplots(
            rows=4,
            cols=1,
            subplot_titles=["Observed", "Trend", "Seasonal", "Residuals"],
        )
        .add_trace(
            go.Scatter(x=x_values, y=result.observed, mode="lines", name='Observed'),
            row=1,
            col=1,
        )
        .add_trace(
            go.Scatter(x=x_values, y=result.trend, mode="lines", name='Trend'),
            row=2,
            col=1,
        )
        .add_trace(
            go.Scatter(x=x_values, y=result.seasonal, mode="lines", name='Seasonal'),
            row=3,
            col=1,
        )
        .add_trace(
            go.Scatter(x=x_values, y=result.resid, mode="lines", name='Residual'),
            row=4,
            col=1,
        )
        .update_layout(
            height=900, title=f'<b>{title}</b>', margin={'t':100}, title_x=0.5, showlegend=False
        )
    )
```

```python
fig = plot_seasonal_decompose(result, dates=train.index)
fig.show()
```

As we can see in the graph above, there is no seasonality in our data

# XGBoost for feature importance and selection

Extreme Gradient Boosting (XGBoost) provides parallel tree boosting. Similar to Random Forests XGBoost also consists of multiple decision trees.<br>The term “gradient boosting” comes from the idea of “boosting” or improving a single weak model by combining it with a number of other weak models in order to generate a collectively strong model. Gradient boosting is an extension of boosting where the process of additively generating weak models is formalized as a gradient descent algorithm over an objective function.
###### Advantages
- **Regularization**:XGBoost has in-built L1 (Lasso Regression) and L2 (Ridge Regression) regularization which prevents the model from overfitting. That is why, XGBoost is also called regularized form of GBM (Gradient Boosting Machine).<br>
- **Handling Missing Values**: XGBoost has an in-built capability to handle missing values. When XGBoost encounters a missing value at a node, it tries both the left and right hand split and learns the way leading to higher loss for each node. It then does the same when working on the testing data.<br>
- **Parallel Processing**: XGBoost utilizes the power of parallel processing and that is why it is much faster than GBM. It uses multiple CPU cores to execute the model.



```python
#seperate predictors from predicted variable
x_train = train.iloc[:,:-1]
y_train = train.iloc[:,-1:]
x_test = test.iloc[:,:-1]
y_test = test.iloc[:,-1:]
```

```python
# fit XGBoost model to our data
#data does not to be sclaed in decision trees
xgb = XGBRegressor(booster ='gbtree',max_depth=3, eval_metric='rmse')
xgb.fit(x_train, y_train)
```

## Auto regressive models

we'll try a different model inputing lags from the variable property prices

First, we check is our data is stationary using the adam fuller test

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=train.index,
        y=train['property_prices'],
        line=dict(color='blue', width=0.7),
        showlegend=False))

fig.update_layout(
        title= 'Property Prices',
        xaxis_title = "Date", 
        yaxis_title = 'Property Prices')

fig.show()
```

```python
p_value = adfuller(train['property_prices'])[1]
print('p-value: ', p_value)
```

Since our p-value is far from 0.5, we fail to reject the null of non-stationarity.<br>
Therefore, we will need to diff the data and check again

```python
p_value_1diff = adfuller(train['property_prices'].diff().iloc[1:])
print('p-value: ', p_value_1diff[1])
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=train.index,
        y=train['property_prices'].diff(),
        line=dict(color='blue', width=0.7),
        showlegend=False))

fig.update_layout(
        title= '1st Diff of Property Prices',
        xaxis_title = "Date", 
        yaxis_title = '1st Diff of Property Prices')

fig.show()
```

Data is still non-stationary

```python
p_value_2diff = adfuller(train['property_prices'].diff().diff().iloc[2:])
print('p-value: ', p_value_2diff[1])
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=train.index,
        y=train['property_prices'].diff().diff(),
        line=dict(color='blue', width=0.7),
        showlegend=False))

fig.update_layout(
        title= '2nd Diff of Property Prices',
        xaxis_title = "Date", 
        yaxis_title = '2nd Diff of Property Prices')

fig.show()
```

```python
train['second_order_property_prices'] = train['property_prices'].diff().diff()
```

Lest's check how many lags to use resosrting to Autocorrelation and Partial Autocorrelation<br>
**Autocorrelation** is a calculation of the correlation of the time series observations with values of the same series, but at previous times.<br>
**Partial Autocorrelation**, on the other hand, summarizes the relationship between an observation in a time series with observations at previous time steps, but with the relationships of intervening observations removed.

```python
# plot autocorrelation
plot_acf(train['second_order_property_prices'].iloc[2:]);
# plot partial autocorrelation
plot_pacf(train['second_order_property_prices'].iloc[2:]);
```

```python
arima_model = ARIMA(train['property_prices'], order = (3,2,0));
arima = arima_model.fit()
print(arima.summary())
```

```python
#predicted values
arima_model_property_prices = pd.Series(arima.forecast(10)[0], index=test.index)

#confidence intervals
lower_ci = [ls[0] for ls in arima.forecast(10)[2]]
upper_ci = [ls[1] for ls in arima.forecast(10)[2]]
```

```python
# plot predicted prices

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df_cleaned.index,
        y=df_cleaned['property_prices'],
        name= 'Real Property Prices',
        line=dict(color='blue', width=0.7)))

# add predicted housing prices
fig.add_trace(
    go.Scatter(
    x=arima_model_property_prices.index,
    y=arima_model_property_prices,
    name= 'Predicted Property Prices',
    mode='lines',
    line=dict(color='red', width=0.7)))

fig.add_trace(
    go.Scatter(
        name='95% CI Upper',
        x=arima_model_property_prices.index,
        y=upper_ci,
        mode='lines',
        marker=dict(color='#444'),
        line=dict(width=0),
        showlegend=False))

fig.add_trace(
go.Scatter(
    name='95% Confidence Interval',
    x=arima_model_property_prices.index,
    y=lower_ci,
    marker=dict(color='#444'),
    line=dict(width=0),
    mode='lines',
    fillcolor='rgba(68, 68, 68, 0.3)',
    fill='tonexty'))

fig.update_layout(
        title= 'Real property Prices vs Predicsted by ARIMA',
        xaxis_title = "Date", 
        yaxis_title = 'Value')

fig.show()
```

```python
# RMSE Computation
ar1_rmse = np.sqrt(MSE(test['property_prices'], arima_model_property_prices))
print("RMSE : % f" %(ar1_rmse))
```
